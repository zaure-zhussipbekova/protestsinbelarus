```python
def select_top_n_dates_with_lag(data, k=5, lag=10):
    top_dates = data.reset_index().sort_values(by='message_text', ascending=False)
    top_dates = top_dates.message_date.values
    ans = [top_dates[0]]
    k -= 1
    for i in range(len(top_dates) - 1):
        if all([abs(d - top_dates[i]).days > lag for d in ans]):
            ans.append(top_dates[i])
            k -= 1
        if k == 0:
            return ans
    return sorted(ans)
```

```python
specific_days = select_top_n_dates_with_lag(data, k = 5, lag = 15)
```

```python
news_channels_df = data
news_channels_df.drop(columns = {'Unnamed: 0'}, inplace = True)
news_channels_df['message_date'] = pd.to_datetime(news_channels_df['message_date'], dayfirst=True)
news_channels_df
```

```python

news_channels_aug = news_channels_df[(news_channels_df['message_date']>='2020-08-01') & (news_channels_df['message_date']<='2020-08-31')]

news_channels_aug['message_date'] = news_channels_aug['message_date'].dt.strftime('%d-%m-%Y')

news_channels_aug = news_channels_aug[['message_date', 'message_text']]

news_channels_aug = news_channels_aug.groupby('message_date').count()
news_channels_aug
```

```python
import matplotlib.pyplot as plt
#local_chats_aug_sep_dates.plot(legend=False, color='#4169E1', linewidth=2.0)

    
plt.xlabel('Date', fontsize=20)

plt.ylabel('Number of posts in news channels', fontsize=20)
dtFmt = mdates.DateFormatter('%d %m') # define the formatting
#plt.gca().xaxis.set_major_formatter(dtFmt)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)




for tick in plt.gca().get_xticklabels():    
    
    tick.set_rotation(90)

#plt.savefig('Number-of-posts-per-day-local-chats.pdf')


pl= plt.bar(x = news_channels_aug.index, height = news_channels_aug.message_text)
for bar in pl:
    plt.annotate(bar.get_height(), 
                 xy=(bar.get_x()+0.07, bar.get_height()+10), 
                     fontsize=10)
plt.show()
```
